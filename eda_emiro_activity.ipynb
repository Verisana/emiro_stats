```python
import pandas as pd
import plotly.express as px
```

```python
filename = 'emiro_activity'
df = pd.read_pickle(filename)
df = pd.DataFrame(df)
df
```

```python
df['weekday'] = df['date_added'].map(lambda x: x.dayofweek)
df['time'] = df['date_added'].map(lambda x: x.time())
df.sort_values(by='time', inplace=True, ascending=False)
df
```

```python
fig = px.scatter(df, x='weekday', y='time')
fig.show()
```
